```python
records = []
for i in range(20):
    neo = x['near_earth_objects'][i]
    record = {
        'Asteroid ID': neo['id'],
        'Name': neo['name'],
        'Absolute Magnitude': neo['absolute_magnitude_h'],
        'Min Diameter (km)': neo['estimated_diameter']['kilometers']['estimated_diameter_min'],
        'Relative Velocity (km/s)': (
            float(neo['close_approach_data'][0]['relative_velocity']['kilometers_per_second'])
            if neo['close_approach_data'] else None
        )
    }
    records.append(record)

for x in records:
  print(x)
```

```python
df = pd.DataFrame(records)
df.to_csv('Asteroids - NeoW.csv')
```
