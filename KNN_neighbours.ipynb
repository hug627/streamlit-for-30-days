```python
from sklearn.datasets import load_iris
import pandas as pd
iris_data = load_iris()
iris = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
display(iris)
```

```python
X_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
display(X_df)
```

```python
y = pd.DataFrame(data=iris_data.target, columns=['target'])
display(y)
```

```python
display(X_df.describe())
```

```python
display(X_df)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_df, y, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_iris = KNeighborsClassifier(n_neighbors = 10)
knn_iris.fit(X_train, y_train)
y_pred = knn_iris.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay,precision_score,recall_score,f1_score,classification_report
```

```python
from sklearn.metrics import confusion_matrix
cn = confusion_matrix(y_test, y_pred)
print(cn)
```

```python
from sklearn import metrics
from matplotlib import pyplot as plt
disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cn, display_labels=iris_data.target_names)
disp.plot()
plt.show()
```

```python
precision_score(y_test, y_pred, average='weighted')
```

```python
recall_score(y_test, y_pred, average='weighted')
```

```python
print(classification_report(y_test,y_pred,target_names=iris_data.target_names))
```

```python
#import the relevant libraries
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

x_train,x_test,y_train,y_test=train_test_split(X_df,y,test_size=0.2,random_state=30) #split our data with test size of 20%

knn=KNeighborsClassifier(n_neighbors=20) #build our knn classifier
knn.fit(x_train,y_train.values.ravel()) #Training KNN classifier
y_pred=knn.predict(x_test)  #Testing
print('Acuuracy=',accuracy_score(y_pred,y_test))
```
